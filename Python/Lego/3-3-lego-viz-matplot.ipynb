```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/lego-cleaned/lego_cleaned.csv") 
```

```python
df.info()
```

```python
print(pd.concat([df.dtypes, df.isnull().sum()], axis=1, keys=['Data Type', 'Null Values']))
```

### **Bar Charts**


* **Number of Lego Sets by Year**

```python
#Calculate the number of sets per year.

sets_per_year = df['year'].value_counts().sort_index()
```

```python
#create the bar chart

 
plt.figure(figsize=(10, 6))
plt.bar(sets_per_year.index, sets_per_year.values, color='orange')


# Add labels and title
plt.xlabel('YEAR')
plt.ylabel('# SETS')
plt.title('# Lego Sets Launched By Year')

# Show the chart
plt.xticks(rotation=75)  # Rotate x-axis labels for better readability
plt.show()
```

* **Number of Sets by Category**

```python
# Count the number of sets per category
sets_per_category = df['category'].value_counts()

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(sets_per_category.index, sets_per_category.values, color='green')

# Add labels and title
plt.xlabel('Category')  # Y-axis now represents years
plt.ylabel('Number of LEGO Sets')
plt.title('# Lego Sets By Category')

```

## Pie Charts

* **Distribution of sets by Theme Group***

```python
# Count the number of sets per category
countsby_category = df['category'].value_counts()
 
```

```python

# Create pie chart
plt.figure(figsize=(5, 5))
plt.pie(countsby_category, labels= countsby_category.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)

# Add title
plt.title('Number of Sets by Category')

# Show the chart
plt.show()
```

# Scatterplot 
*  **Correlation between the number of Lego pieces a nd the price of the sets**

```python
import matplotlib.pyplot as plt

# Scatter plot: Price vs. Number of Pieces
plt.figure(figsize=(12,6))
plt.scatter(df['pieces'], df['US_retailPrice'], color='red', alpha=0.4)

# Add labels and title
plt.xlabel('# Pieces/Set')
plt.ylabel('Retail Price (USD)')
plt.title('Price vs. # of Pieces LEGO Sets')

# Show the chart
plt.show()
```

## Line Chart

*  **Show the trend of price of lego sets through the years**

```python
import matplotlib.pyplot as plt

# Calculate the average price per year
avg_price_per_year = df.groupby('year')['US_retailPrice'].mean()
```

```python
# Create line chart
plt.figure(figsize=(12, 5))
plt.plot(avg_price_per_year.index, avg_price_per_year.values, marker='o', color='blue')


# Add labels and title

plt.xlabel('Year')
plt.ylabel('Average Price (USD)')
plt.title('Average LEGO Set Price per Year')

# Display chart

plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
