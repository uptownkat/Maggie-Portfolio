```python
#view side by side

print(pd.concat([df.dtypes, df.isnull().sum()], axis=1, keys=['Data Type', 'Null Values']))
```

```python
# Average age  
avg_age = df['agerange_min'].dropna().mean()
```

```python
print(avg_age)
```

#### **Average Price:**

```python
avg_price = df['US_retailPrice'].mean()
```

```python
print(avg_price)
```

#### **# of Lego Sets:**

#### **The earliest and latest years:**

```python

earliest_year = df['year'].min()
latest_year = df['year'].max()

print(f"Earliest year: {earliest_year}")
print(f"Most recent year: {latest_year}")
```

#### **Most Expensive Lego Set:**

```python

# Pull the row with the highest US retail price.

most_expensive_set = df.loc[df['US_retailPrice'].idxmax()]

 
print("Most Expensive LEGO Set:")
print(most_expensive_set)
```

#### ***Least Expensive Lego Set:**

```python

# Pull the row with the lowest price

least_expensive_set = df.loc[df['US_retailPrice'].idxmin()]

print("Least Expensive LEGO Set:")
print(least_expensive_set)
```

#### **Set with the most # pieces:**

```python
# Pull the row that has the maximum # of pieces

most_pieces_set = df.loc[df['pieces'].idxmax()]

 
print("LEGO Set with the Most Pieces:")
print(most_pieces_set)
```

#### **Set with the least # pieces:**

```python
#Pull row with the smallest number of pieces

least_pieces_set = df.loc[df['pieces'].idxmin()]

 
print("LEGO Set with the Least Pieces:")
print(least_pieces_set)
```
