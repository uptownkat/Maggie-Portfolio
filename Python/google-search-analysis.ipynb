```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pytrends.request import TrendReq

# Initialize Pytrends
pytrend = TrendReq()

# Define keyword and get interest by region
keyword = "Generative AI"
pytrend.build_payload(kw_list=[keyword], timeframe="today 5-y")  # Last 5 years
data = pytrend.interest_by_region()

# Reset index for proper formatting
data = data.reset_index()

# Sort data by search interest
data = data.sort_values(by=keyword, ascending=False).head(10)  # Top 10 countries

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(data.set_index("geoName"), cmap="Blues", annot=True, fmt=".0f", linewidths=0.5)

# Title and labels
plt.title(f"Google Search Interest in '{keyword}' by Country", fontsize=14)
plt.xlabel("Country")
plt.ylabel("Search Interest")

# Display the heatmap
plt.show()
```

```python
# Plot Line chart

# Set keyword and timeframe (last 5 years)
keyword = ["Generative AI"]
pytrend.build_payload(keyword, timeframe="2019-02-13 2024-02-13", geo="")  # Global search

# Get interest over time
data = pytrend.interest_over_time()

# Drop 'isPartial' column if it exists
if 'isPartial' in data.columns:
    data = data.drop(columns=['isPartial'])

# Plot the line chart
plt.figure(figsize=(12,6))
plt.plot(data.index, data["Generative AI"],linestyle="-", color="red", label="Generative AI Search Interest")

# Formatting the plot
plt.xlabel("Year", fontsize=12)
plt.ylabel("Search Interest", fontsize=12)
plt.title("Google Search Trend for 'Generative AI' (Last 5 Years)", fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq

# Initialize Pytrends
pytrend = TrendReq()

# Define keyword and get interest by region
keyword = "Generative AI"
pytrend.build_payload(kw_list=[keyword], timeframe="today 5-y")  # Last 5 years
data = pytrend.interest_by_region()

# Reset index for formatting
data = data.reset_index()

# Get the top 10 countries with the highest search interest
top_countries = data.sort_values(by=keyword, ascending=False).head(5)

# Pie chart
plt.figure(figsize=(10, 6))
plt.pie(
    top_countries[keyword], 
    labels=top_countries["geoName"], 
    autopct="%1.1f%%", 
    colors=plt.cm.Paired.colors, 
    startangle=140
)

# Title
plt.title(f"Search Interest in '{keyword}' by Country (Top 5)")

# Display the pie chart
plt.show()
```
