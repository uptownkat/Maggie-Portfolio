## **Line Chart**

(Line charts show progression over time)

**A. Plotting Dots on Chart**

    x=[1,4,6]
    y=[5,2,7]


**B. Displaying Lines on Chart**

    plt.plot(x,y)


**C. Display Labels on X and Y axis**

    plt.xlabel('X Label')
    plt.ylabel('Y Label')


**D. Display a Title**

    plt.title('Title')


**E. Display Chart**

    plt.show()

```python
Day = ['Monday', 'Tuesday','Wednesday','Thursday', 'Friday','Saturday','Sunday']
Temperature = [63,78,90,65,46,74,88]

plt.plot(Day, Temperature)

plt.xlabel ('Day')
plt.ylabel ('Temp (F)')

plt.title ('Weekly Temperature')
plt.show()
```
